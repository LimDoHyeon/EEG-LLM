# JSON to JSONL

```python
import json
# JSONL  
train_jsonl_file_path = '/Users/imdohyeon/Documents/PythonWorkspace/EEG-GPT/Dataset/jsonl/train.jsonl'
test_jsonl_file_path = '/Users/imdohyeon/Documents/PythonWorkspace/EEG-GPT/Dataset/jsonl/test.jsonl'
```

```python
# JSON   
def load_json(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return json.load(file)

# JSONL    (completion   )
def save_to_jsonl(data, file_path):
    with open(file_path, 'w', encoding='utf-8') as jsonl_file:
        for entry in data:
            # completion   
            entry['completion'] = str(entry['completion'])
            json.dump(entry, jsonl_file)
            jsonl_file.write('\n')

# JSON  
train_data = load_json(train_output_path)
test_data = load_json(test_output_path)

# JSON  JSONL  
save_to_jsonl(train_data, train_jsonl_file_path)
save_to_jsonl(test_data, test_jsonl_file_path)

print(f"Converted {train_output_path} to {train_jsonl_file_path}")
print(f"Converted {test_output_path} to {test_jsonl_file_path}")
```
